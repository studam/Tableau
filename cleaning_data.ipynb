```python
# Dependencies & Setup
#--------------------------
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# Files
#------------------------------------------
jun2018 = "201806-citibike-tripdata.csv"
jul2018 = "201807-citibike-tripdata.csv"
aug2018 = "201808-citibike-tripdata.csv"
jun2021 = "JC-202106-citibike-tripdata.csv"
jul2021 = "JC-202107-citibike-tripdata.csv"
aug2021 = "JC-202108-citibike-tripdata.csv"
dec2021 = "JC-202112-citibike-tripdata.csv"
jan2022 = "JC-202201-citibike-tripdata.csv"
feb2022 = "JC-202202-citibike-tripdata.csv"
```

```python
# Read Data 2018, 2021 & 2022 Citi Bike Data Files & Store into Pandas DataFrames
#------------------------------------------------------------------------------------

jun2018_df = pd.read_csv(jun2018, encoding="ISO-8859-1")
jul2018_df = pd.read_csv(jul2018, encoding="ISO-8859-1")
aug2018_df = pd.read_csv(aug2018, encoding="ISO-8859-1")
jun2021_df = pd.read_csv(jun2021, encoding="ISO-8859-1")
jul2021_df = pd.read_csv(jul2021, encoding="ISO-8859-1")
aug2021_df = pd.read_csv(aug2021, encoding="ISO-8859-1")
dec2021_df = pd.read_csv(dec2021, encoding="ISO-8859-1")
jan2022_df = pd.read_csv(jan2022, encoding="ISO-8859-1")
feb2022_df = pd.read_csv(feb2022, encoding="ISO-8859-1")
```

```python
# Combine All DataFrames Into Single DataFrame
#---------------------------------------------------------------------------
summer2018_df = jun2018_df.append([jul2018_df, aug2018_df], ignore_index=True)
summer2018_df.head()
```

```python
# Combine All DataFrames Into Single DataFrame
#---------------------------------------------------------------------------
summer2021_df = jun2021_df.append([jul2021_df, aug2021_df], ignore_index=True)
summer2021_df.head()
```

```python
# Combine All DataFrames Into Single DataFrame
#--------------------------------------------------------------------------
winter2021_df = dec2021_df.append([jan2022_df, feb2022_df], ignore_index=True)
winter2021_df.head()
```

```python
# Convert All DateTimes to "%Y-%m-%d %H:%M:%S" Format
#------------------------------------------------------------------------
summer2018_df["starttime"] = pd.to_datetime(summer2018_df["starttime"])
summer2018_df["stoptime"] = pd.to_datetime(summer2018_df["stoptime"])
summer2018_df.head()
```

```python
# Convert All DateTimes to "%Y-%m-%d %H:%M:%S" Format
#--------------------------------------------------------------------------

summer2021_df["started_at"] = pd.to_datetime(summer2021_df["started_at"])
summer2021_df["ended_at"] = pd.to_datetime(summer2021_df["ended_at"])
summer2021_df.head()
```

```python
# Convert All DateTimes to "%Y-%m-%d %H:%M:%S" Format
#--------------------------------------------------------------------------
winter2021_df["started_at"] = pd.to_datetime(winter2021_df["started_at"])
winter2021_df["ended_at"] = pd.to_datetime(winter2021_df["ended_at"])
winter2021_df.head()
```

```python
# Save Combined DataFrames to CSV
#---------------------------------------------------------------
summer2018_df.to_csv("Citi_summer2018.csv", index=False, header=True)
summer2021_df.to_csv("Citi_summer2021.csv", index=False, header=True)
winter2021_df.to_csv("Citi_winter2021.csv", index=False, header=True)
```
